```python
# 线性回归算法
# 1、损失函数，如平方损失
# 2、优化算法
#
import torch
import time
import numpy as np
import matplotlib.pyplot as plt
import data_dj
```

```python
# 自己的包文件data_dj.py在该文件的当前目录下
features, labels = data_dj.create_data(1000, 2)
print(features[0], labels[0])
```

```python
data_dj.showData(features[:, 1], labels)
```

```python
# 模型训练过程
batch_size = 10
w = torch.tensor(np.random.normal(0, 0.01, (2, 1)),
                     dtype=torch.float32, requires_grad=True)
b = torch.zeros(1, dtype=torch.float32, requires_grad=True)
lr = 0.03
num_epochs = 3
net = data_dj.linreg
loss = data_dj.squared_loss
for epoch in range(num_epochs):
    for x, y in data_dj.data_iter(batch_size, features, labels):
        # 通过模型的前向传播计算预测值，然后利用损失函数计算与真实值之间的差距，
        # 然后对损失函数进行反向传播，利用优化函数更新权重等参数；最后需要将参数的导数归0
        l = loss(net(x, w, b), y).sum()
        l.backward()
        data_dj.sgd([w, b], lr, batch_size)
        w.grad.data.zero_()
        b.grad.data.zero_()
    train_l = loss(net(features, w, b), labels)
    print('epoch %d, loss %f' % (epoch + 1, train_l.mean().item()))
print(w, b)
```
